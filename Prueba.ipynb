```python
import pandas as pd
```

```python
df = pd.read_csv(r'movies_dataset.csv')
df.head(2)
```

```python
df[['revenue', 'budget']] = df[['revenue', 'budget']].fillna(0)
```

```python
df = df.dropna(subset=['release_date'])
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')
df['release_year'] = df['release_date'].dt.year.fillna(0).astype(int)
```

```python
# Convierte las columnas 'budget' y 'revenue' a numéricas para poder crear la columna solicitada
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

#Creamos la nueva columna return
df['return'] = df['revenue'] / df['budget']
df['return'] = df['return'].fillna(0)
```

```python
df = df.drop(['video', 'imdb_id', 'adult', 'original_title', 'vote_count', 'poster_path', 'homepage'], axis=1)
```

```python
#Creamos la columna 'day_of_week'
# Obtener el día de la semana como un entero
day_of_week_int = df['release_date'].dt.dayofweek

# Asignar el número de día de la semana con su respectivo string
day_of_week_dict = {
    0: 'lunes',
    1: 'martes',
    2: 'miércoles',
    3: 'jueves',
    4: 'viernes',
    5: 'sábado',
    6: 'domingo'
}

# Utilizar el diccionario para obtener el string correspondiente al día de la semana
df['day_of_week'] = day_of_week_int.map(day_of_week_dict)
```

```python
import pandas as pd
import re

def extraer_datos(df, columna):
    # Función para extraer los valores de una cadena, lista o diccionario
    def extraer_valores(valor):
        if isinstance(valor, str):
            # Si es una cadena, extraer los valores con la expresión regular
            pattern = r"'name': '([^']*)'"
            coincidencias = re.findall(pattern, valor)
            return coincidencias
        elif isinstance(valor, list):
            # Si es una lista, extraer los valores de cada elemento de la lista
            valores = []
            for elemento in valor:
                if 'name' in elemento:
                    valores.append(elemento['name'])
            return valores
        elif isinstance(valor, dict):
            # Si es un diccionario, extraer los valores de la clave 'name'
            if 'name' in valor:
                return valor['name']
        return None
    
    # Aplicar la función a la columna y sobrescribir la columna
    df[columna] = df[columna].apply(extraer_valores)
    
    return df
   
```

```python
def retorno(pelicula):
    # Filtro la información de la película en base a su título
    pelicula_info = df_p[df_p['title'] == pelicula].iloc[0]
    
    # Obtengo la inversión, ganancia, retorno y año de lanzamiento de la película
    inversion = pelicula_info['budget']
    ganancia = pelicula_info['revenue']
    retorno = pelicula_info['return']
    anio = pelicula_info['release_year']
    
    # Retorno la información en un diccionario
    return {'pelicula': pelicula, 'inversion': inversion, 'ganancia': ganancia, 'retorno': retorno, 'anio': anio}
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_recommendation(titulo):
    # Obtener la fila correspondiente al título de la película
    idx = df_p[df_p['title'] == titulo].index[0]
    
    # Obtener la matriz de características
    features = df_p.iloc[:, 7:19]
    
    # Calcular la matriz de similitud de cosenos
    sim_matrix = cosine_similarity(features)
    
    # Obtener los puntajes de similaridad para la película de interés
    sim_scores = list(enumerate(sim_matrix[idx]))
    
    # Ordenar las películas según su puntaje de similaridad
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Obtener las 5 películas más similares como recomendaciones
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    recommendation = df_p['title'].iloc[movie_indices].tolist()
    
    # Retornar la lista de recomendaciones
    return {'lista recomendada': recommendation}
```
